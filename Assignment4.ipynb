```python
### function to find the best hyperparameter for the decision tree and will used
### to calculate the test error for the testing set
# *********************************Parameters***********************************
# X_train: training set
# X_valid: validation set
# t_train: training set targets
# t_valid: validation set targets

def decision_tree(X_train, X_test, t_train, t_test):
    print("************************* Decision Tree Classifier ******************************")
    
    kf = KFold(n_splits = 5) # assign K = 5 
    kf.get_n_splits(X_train) # split training set
    
    missClass = np.zeros(398) # empty array for missclassification rate
    j = 0
    
    # looping through the possible hyperparameter options
    for k in range (2, 400, 1):
        #classifier = 0
        classifier = tree.DecisionTreeClassifier(max_leaf_nodes = k, random_state = 2455) # creating the decision tree classifier
        
        # KFold splitting the the training set
        for train_i, test_i in kf.split(X_train):
            x_train, x_test = X_train[train_i], X_train[test_i]
            y_train, y_test = t_train[train_i], t_train[test_i]

            N = len(x_train)
            M = len(x_test)

            #*******Debugging print statementes**********
            #print(x_train.shape)
            #print(x_test.shape)
            #print("Training set:",train_i)
            #print("Test set:",test_i)

            True_pos = 0 # variable to hold the value for the True positive count
            False_pos = 0 # variable to hold the value for the False positive count
            True_neg = 0 # variable to hold the value for the True negative count
            False_neg = 0 # varible to hold the value fro the False negative count

            classifier = classifier.fit(x_train, y_train) # training the classifier
 
            #prediction = classifier.predict(x_test) # using the classifier to predict 

            # finding the true positives, true negatives, false positives, false, negatives
            #for i in range(M):
            #    if(prediction[i] == 1 and y_test[i] == 1):
            #        True_pos += 1
            #    elif(prediction[i] == 0 and y_test[i] == 0):
            #        True_neg += 1
            #    elif(prediction[i] > y_test[i]):
            #        False_pos += 1
            #    elif(prediction[i] < y_test[i]):
            #        False_neg += 1

            #missClass[j] += (False_neg+False_pos)/(True_pos+False_pos+True_neg+False_neg) # calculating the missclassification rate for each hyperparameter
            missClass[j] += 1 - classifier.score(x_test, y_test)
            #print("Prediction:", prediction[:20])
        j += 1 # iterating through the missClass array
            
    missClass /= 5 # finding the cross-validation error for each hyperparameter
    best_hyperparameter = np.argmin(missClass) # finding the smallest hyperparameter
    
    x_axis = np.linspace(2, 400, num = len(missClass))
    #y_axis = np.full((1,399), missClass)
    
    plt.scatter(x_axis, missClass, color = 'red')
    plt.title('Validation Error - Decision Tree')
    plt.ylabel('Validation Error')
    plt.xlabel('Max number of leaves')
    plt.show()
    
    #*******Debugging print statementes**********
    print("Cross-validation errors:",missClass) 
    #print(missClass.shape)
    print("Best maximum number of leaves:", best_hyperparameter)
    #print("Prediction:", prediction[:25])
        
    #tree.plot_tree(classifer)
    
    N = len(X_train)
    M = len(X_test)
    
    True_pos = 0
    True_neg = 0
    False_pos = 0
    False_neg = 0
    
    classifier = tree.DecisionTreeClassifier(max_leaf_nodes = best_hyperparameter, random_state = 2455)
    classifier = classifier.fit(X_train, t_train) # training the classifier using the training set
    
    #prediction = classifier.predict(X_test) # using test set with the classifier to predict
    
    # finding the true positives, true negatives, false positives, false, negatives
    #for i in range(M):
    #    if(prediction[i] == 1 and t_test[i] == 1):
    #        True_pos += 1
    #    elif(prediction[i] == 0 and t_test[i] == 0):
    #        True_neg += 1
    #    elif(prediction[i] > t_test[i]):
    #        False_pos += 1
    #    elif(prediction[i] < t_test[i]):
    #        False_neg += 1
    
    #test_missClass = (False_neg+False_pos)/(True_pos+False_pos+True_neg+False_neg) # calculating the missclassification rate for each hyperparameter
    test_missClass = 1 - classifier.score(X_test, t_test)
    
    #*******Debugging print statementes**********    
    print("Test error:", test_missClass)
    
    x_axis = np.linspace(2, 400, num = 399)
    y_axis = np.full((1,399), test_missClass)
    
    #*******Debugging print statementes**********
    #print(x_axis)
    #print(x_axis.shape)
    #print(y_axis[0])
    #print(y_axis.shape)
    
    plt.scatter(x_axis, y_axis[0], color = 'red')
    plt.title('Test Error - Decision Tree')
    plt.ylabel('Test Error')
    plt.xlabel('Max number of leaves')
    plt.show()
    
    print("END OF DECISION TREE")
    
    return test_missClass
```
